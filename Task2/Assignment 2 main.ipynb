```python
from ultralytics import YOLO

# Load the trained model (you can replace with your custom model path if needed)
model = YOLO('yolov8n.pt')  # or 'runs/detect/train/weights/best.pt' if you've trained your own

results = model(r"C:\Users\anees\OneDrive\Pictures\Images\road.jpeg")
results[0].show()


# Print detection details
for box in results[0].boxes:
    cls_id = int(box.cls[0])
    conf = float(box.conf[0])
    coords = box.xyxy[0].tolist()
    print(f"Class ID: {cls_id}, Confidence: {conf:.2f}, Box: {coords}")
```
