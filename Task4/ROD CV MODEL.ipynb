```python
import os

folder_path = r"C:\Users\anees\OneDrive\Desktop\IIIT H\f\rod-dataset\images\test"

# Get image files
images = [f for f in os.listdir(folder_path)
          if f.lower().endswith(('.jpg', '.png', '.jpeg'))]

# Sort to keep order consistent
images.sort()

# Rename images
for i, img in enumerate(images, start=1):
    ext = os.path.splitext(img)[1]  # keep original extension
    new_name = f"test{i}{ext}"
    
    os.rename(
        os.path.join(folder_path, img),
        os.path.join(folder_path, new_name)
    )

print("✅ Images renamed successfully")
```

```python
from ultralytics import YOLO
import cv2

# Load trained model
model = YOLO(r"C:\Users\anees\runs\detect\train10\weights\best.pt")

# Read image
img_path = r"C:\Users\anees\OneDrive\Desktop\IIIT H\f\rod-dataset\images\test\test1.jpg"
img = cv2.imread(img_path)

# Run detection
results = model(img, conf=0.4)

# Count rods
rod_count = len(results[0].boxes)

# Draw bounding boxes
annotated_img = results[0].plot()

# Put count text on image
cv2.putText(
    annotated_img,
    f"Rod Count: {rod_count}",
    (20, 40),
    cv2.FONT_HERSHEY_SIMPLEX,
    1,
    (0, 255, 0),
    2
)

# Save output
output_path = "rod_count_output.jpg"
cv2.imwrite(output_path, annotated_img)

print("✅ Rod count:", rod_count)
print("✅ Saved image as:", output_path)
```

```python
from ultralytics import YOLO
import cv2
import os

# Load trained YOLO model
model = YOLO(r"C:\Users\anees\runs\detect\train10\weights\best.pt")

# Input test images folder
input_folder = r"C:\Users\anees\OneDrive\Desktop\IIIT H\f\rod-dataset\images\test"

# Output folder on Desktop
output_folder = r"C:\Users\anees\OneDrive\Desktop\test"

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Process all images in test folder
for img_name in os.listdir(input_folder):
    if img_name.lower().endswith(('.jpg', '.png', '.jpeg')):
        img_path = os.path.join(input_folder, img_name)

        # Read image
        img = cv2.imread(img_path)

        # Run detection
        results = model(img, conf=0.4)

        # Count rods
        rod_count = len(results[0].boxes)

        # Draw bounding boxes
        annotated_img = results[0].plot()

        # Put rod count text on image
        cv2.putText(
            annotated_img,
            f"Rod Count: {rod_count}",
            (20, 40),
            cv2.FONT_HERSHEY_SIMPLEX,
            1,
            (0, 255, 0),
            2
        )

        # Save output image
        output_path = os.path.join(output_folder, img_name)
        cv2.imwrite(output_path, annotated_img)

        print(f"{img_name} → Rod Count: {rod_count}")

print("✅ All test images processed and saved on Desktop/test")
```
