```python
from pathlib import Path
import subprocess, sys, os
from tqdm import tqdm

WORK_DIR = Path("video_processing")   
WORK_DIR.mkdir(exist_ok=True)

# ðŸŸ© change this to your downloaded video filename
VIDEO_FILE = Path(r"C:\Users\anees\input\original_video.f399.mp4")   # e.g. Path("sample.mp4")

FRAMES_DIR = WORK_DIR / "frames"
FRAMES_DIR.mkdir(exist_ok=True)

PROCESSED_DIR = WORK_DIR / "processed"
PROCESSED_DIR.mkdir(exist_ok=True)

FPS = 24   # how many frames per second to extract
```

```python
# extract frames from video at 24 FPS
frame_pattern = "frame_%06d.png"
cmd = f'ffmpeg -hide_banner -loglevel error -i "{VIDEO_FILE}" -vf fps={FPS} "{FRAMES_DIR/frame_pattern}"'
print(cmd)
subprocess.run(cmd, shell=True, check=True)
print("âœ… Frames saved in:", FRAMES_DIR)
```

```python
from ultralytics import YOLO

# Choose model:
#   yolov8n.pt = object detection
#   yolov8n-seg.pt = segmentation (masks)
model = YOLO("yolov8n.pt")   # downloads weights automatically

frame_files = sorted(FRAMES_DIR.glob("*.png"))
print("Total frames:", len(frame_files))

# process frames one-by-one (CPU-friendly)
for f in tqdm(frame_files, desc="Processing frames with YOLOv8"):
    model.predict(
        source=str(f),
        device="cpu",       # force CPU mode
        imgsz=640,          # smaller size â†’ faster
        conf=0.25,
        save=True,
        save_txt=False,
        save_dir=str(PROCESSED_DIR)
    )

print("âœ… All processed images are saved in:", PROCESSED_DIR)
```
